```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
csvpath = Path("UFOs_coord.csv")
```

```python
ufo_data = pd.read_csv(csvpath,encoding="latin-1")
ufo_data.head()
```

```python
ufo_data[["Date", "Time"]] = ufo_data["Date / Time"].str.split(" ", expand=True)
ufo_data.head(100)
```

```python
ufo_data = ufo_data.drop(columns=["Date / Time", "Time"])
ufo_data.head()
```

```python
from dateutil.parser import parse
new_dates = []
for i in range(len(ufo_data)):
    new_date = parse(ufo_data.loc[i,"Date"])
    new_dates.append(new_date)
new_dates
```

```python
ufo_data["Clean Date"] = new_dates
ufo_data
```

```python
ufo_data = ufo_data.drop(columns=["Date"])
```

```python
ufo_data.to_csv("clean_UFO_data.csv")
```
