```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
csvpath = Path("UFOs_coord.csv")
```

```python
ufo_data = pd.read_csv(csvpath,encoding="latin-1")
ufo_data.head()
```

```python
ufo_data[["Date", "Time"]] = ufo_data["Date / Time"].str.split(" ", expand=True)
ufo_data.head(100)
```

```python
ufo_data = ufo_data.drop(columns=["Date / Time", "Time"])
ufo_data.head()
```

```python
ufo_data["Date"] = pd.to_datetime(ufo_data["Date"])
ufo_data.head(100)
```

```python
ufo_data[["Year", "Month/Day"]] = ufo_data["Date"].str.split("-", expand=True)
ufo_data.head(100)
```

```python
ufo_data["Date"] = ufo_data["Date"].dt.strftime("%Y%m%d")
ufo_data.head(100)
```
